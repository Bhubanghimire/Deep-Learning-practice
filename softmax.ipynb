```python
import matplotlib.pyplot as plt
from keras.activations import softmax
import tensorflow as tf
```

```python
x=[1,2,6]
x_tensor = tf.constant(x,dtype=tf.float32)
```

```python
result = softmax(x_tensor,axis=-1)
```

```python
plt.bar(range(len(x_tensor)), result)
plt.title("Softmax Function")
plt.xlabel("Index")
plt.ylabel("softmax(x)")
plt.xticks(range(len(x_tensor)), [f"x{i}" for i in range(len(x_tensor))])
plt.grid(True)
plt.show()
```
