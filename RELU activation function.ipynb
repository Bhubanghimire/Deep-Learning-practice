```python
import matplotlib.pyplot as plt
```

```python
from keras.activations import relu
```

```python
x= [-10, -5,0.0, 5,10]
```

```python
result =relu(x)
```

```python
result1=relu(x, negative_slope=0)
result2=relu(x, negative_slope=10.5)
result3=relu(x, max_value=1)
result4=relu(x, threshold=2)
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, result1, label='ReLU with negative slope=0', marker='o')

plt.title('ReLU Activation Function with Negative Slope')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(18, 6))
plt.plot(x, result2, label='ReLU with negative slope=0', marker='o')

plt.title('ReLU Activation Function with Negative Slope')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(18, 6))
plt.plot(x, result3, label='ReLU with negative slope=0', marker='o')

plt.title('ReLU Activation Function with Negative Slope')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(18, 6))
plt.plot(x, result4, label='ReLU with negative slope=0', marker='o')

plt.title('ReLU Activation Function with Negative Slope')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
